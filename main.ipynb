```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")

#print(train_df.head())
#print(test_df.head())
X_train = train_df.drop(columns=["rainfall", "id", "day"])
X_test = test_df.drop(columns=["id", "day"])
y_train = train_df["rainfall"]

my_RandomForest = RandomForestClassifier(n_estimators=15, bootstrap = True, random_state=3)
my_RandomForest.fit(X_train, y_train)
rf_pred = my_RandomForest.predict(X_test)
#print(rf_pred)

```
